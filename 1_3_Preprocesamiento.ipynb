Aca con el siguiente codigo realizo la sumatoria de cada transaccion por cliente y lo asigno a una fila y un nuevo data,llegando asi al numero deseado 456369 que son los clientes unicos

```python
df_final = df.groupby('Unique_ID_client').agg({'Unique_ID_client': 'first', 'duracion_meses': 'first', 'product10': 'first', 'pk_partition': 'last', 'entry_date': 'first', 'entry_channel': 'first', 'active_customer': 'last', 'segment': 'first', 'short_term_deposit': 'first', 'loans': 'first', 'mortgage': 'first', 'funds': 'first', 'securities': 'first', 'long_term_deposit': 'first', 'short_term_deposit': 'first', 'Cuenta easyMoney++': 'first', 'credit_card': 'first', 'payroll': 'first', 'pension_plan': 'first', 'payroll_account': 'first', 'Cuenta easyMoney Crypto': 'first', 'debit_card': 'first', 'Cuenta easyMoney+': 'first', 'Cuenta easyMoney': 'first', 'country_id': 'first', 'region_code': 'first', 'gender': 'first', 'age': 'first', 'deceased': 'first', 'salary': 'first', 'product40': 'first', 'product60': 'first', 'dias_entre_compras': 'first', 'compra_real': 'first', 'tiempo_desde_ultima_compra_otro': 'first', 'tiempo_entre_compras_mismo_producto': 'first', 'tiempo_desde_ultima_compra_otro_partition': 'first', 'tiempo_entre_compras_mismo_producto_partition': 'first'})
```

```python
df_final['entry_year'] = df_final['entry_date'].dt.year
df_final['entry_month'] = df_final['entry_date'].dt.month
df_final['entry_day'] = df_final['entry_date'].dt.day
```

```python
df_final['pk_partition_year'] = df_final['pk_partition'].dt.year
df_final['pk_partition_month'] = df_final['pk_partition'].dt.month
df_final['pk_partition_day'] = df_final['pk_partition'].dt.day
```

```python
 #Establecer 'entry_date' como índice y ordenar el DataFrame
df_final.set_index('entry_date', inplace=True)
df_final.sort_index(ascending=True, inplace=True)
```

```python
# Crear un umbral para la cantidad mínima de registros para no ser agrupados como "Otros"
umbral = 200

# Encontrar las categorías que tienen menos de 'umbral' registros
categorias_menos_frecuentes = df_final['entry_channel'].value_counts()[df_final['entry_channel'].value_counts() < umbral].index

# Reemplazar esas categorías con "Otros"
df_final.loc[df_final['entry_channel'].isin(categorias_menos_frecuentes), 'entry_channel'] = 'Otros'

# Imprimir el DataFrame con las categorías agrupadas
print(df_final['entry_channel'].value_counts())
```
