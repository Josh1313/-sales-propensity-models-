```python
df['product10'] = df['product10'].astype(int)
df['product40'] = df['product40'].astype(int)
df['product60'] = df['product60'].astype(int)
```

```python
print(df.columns)
```

```python
# Definimos las columnas de los productos financieros
product10_columns = ['Cuenta easyMoney', 'Cuenta easyMoney+', 'Cuenta easyMoney++', 'Cuenta easyMoney Crypto']
product40_columns = ['funds', 'long_term_deposit', 'pension_plan', 'securities', 'short_term_deposit']
product60_columns = ['credit_card', 'debit_card', 'loans', 'mortgage']

# Calculamos las ganancias por cliente en función de los productos vendidos
df['Total profit by client'] = (df[product10_columns].sum(axis=1) * 10 +
                                df[product40_columns].sum(axis=1) * 40 +
                                df[product60_columns].sum(axis=1) * 60)

# Agrupamos por cliente, mes y año para calcular las ganancias correspondientes
data_grouped1 = df.groupby(['Unique_ID_client', 'entry_year', 'entry_month'])['Total profit by client'].sum().reset_index()

# Calculamos las ganancias por cliente, por mes y por año
data_grouped1['Total profit by M'] = data_grouped1.groupby(['Unique_ID_client', 'entry_year', 'entry_month'])['Total profit by client'].transform('sum')
data_grouped1['Total profit by Y'] = data_grouped1.groupby(['Unique_ID_client', 'entry_year'])['Total profit by client'].transform('sum')


```

```python
data_grouped1
```

```python
product10_columns = ['Cuenta easyMoney', 'Cuenta easyMoney+', 'Cuenta easyMoney++', 'Cuenta easyMoney Crypto']
product40_columns = ['funds', 'long_term_deposit', 'pension_plan', 'securities', 'short_term_deposit']
product60_columns = ['credit_card', 'debit_card', 'loans', 'mortgage']

# Calculamos las ganancias por cliente en función de los productos vendidos
df['Profit by Client(10)'] = df[product10_columns].sum(axis=1) * 10
df['Profit by Client(40)'] = df[product40_columns].sum(axis=1) * 40
df['Profit by Client(60)'] = df[product60_columns].sum(axis=1) * 60

# Agrupamos por cliente, mes y año para calcular las ganancias correspondientes
data_grouped_10 = df.groupby(['Unique_ID_client', 'entry_year', 'entry_month'])['Profit by Client(10)'].sum().reset_index()
data_grouped_40 = df.groupby(['Unique_ID_client', 'entry_year', 'entry_month'])['Profit by Client(40)'].sum().reset_index()
data_grouped_60 = df.groupby(['Unique_ID_client', 'entry_year', 'entry_month'])['Profit by Client(60)'].sum().reset_index()

# Calculamos las ganancias por cliente, por mes y por año
data_grouped_10['Profit by M(10)'] = data_grouped_10.groupby(['Unique_ID_client', 'entry_month'])['Profit by Client(10)'].transform('sum')
data_grouped_10['Profit by Y(10)'] = data_grouped_10.groupby(['Unique_ID_client', 'entry_year'])['Profit by Client(10)'].transform('sum')
data_grouped_40['Profit by M(40)'] = data_grouped_40.groupby(['Unique_ID_client', 'entry_month'])['Profit by Client(40)'].transform('sum')
data_grouped_40['Profit by Y(40)'] = data_grouped_40.groupby(['Unique_ID_client', 'entry_year'])['Profit by Client(40)'].transform('sum')
data_grouped_60['Profit by M(60)'] = data_grouped_60.groupby(['Unique_ID_client', 'entry_month'])['Profit by Client(60)'].transform('sum')
data_grouped_60['Profit by Y(60)'] = data_grouped_60.groupby(['Unique_ID_client', 'entry_year'])['Profit by Client(60)'].transform('sum')
```

```python
# Combina los DataFrames por Unique_ID_client y entry_year, entry_month
merged_df = data_grouped_10.merge(data_grouped_40, on=['Unique_ID_client', 'entry_year', 'entry_month'], how='outer')
merged_df = merged_df.merge(data_grouped_60, on=['Unique_ID_client', 'entry_year', 'entry_month'], how='outer')
merged_df = merged_df.merge(data_grouped1, on=['Unique_ID_client', 'entry_year', 'entry_month'], how='outer')

# Elimina duplicados basados en Unique_ID_client
merged_df = merged_df.drop_duplicates(subset=['Unique_ID_client'])

# Imprime el DataFrame resultante
print(merged_df)

```

```python
# Combina el DataFrame merged_df con el DataFrame original df usando outer join
final_merged_df = df.merge(merged_df, on=['Unique_ID_client', 'entry_year', 'entry_month'], how='outer')

# Elimina duplicados basados en Unique_ID_client
#final_merged_df = final_merged_df.drop_duplicates(subset=['Unique_ID_client'])
```
